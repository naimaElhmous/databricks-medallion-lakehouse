```python
INGESTION_CONFIG = [
    {
        "source": "crm",
        "path": "/Volumes/lakehouse/bronze/raw_sources/engineering/source_crm/cust_info.csv",
        "table": "crm_cust_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/lakehouse/bronze/raw_sources/engineering/source_crm/prd_info.csv",
        "table": "crm_prd_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/lakehouse/bronze/raw_sources/engineering/source_crm/sales_details.csv",
        "table": "crm_sales_details"
    },
    {
        "source": "erp",
        "path": "/Volumes/lakehouse/bronze/raw_sources/engineering/source_erp/CUST_AZ12.csv",
        "table": "erp_cust_az12"
    },
    {
        "source": "erp",
        "path": "/Volumes/lakehouse/bronze/raw_sources/engineering/source_erp/LOC_A101.csv",
        "table": "erp_loc_a101"
    },
    {
        "source": "erp",
        "path": "/Volumes/lakehouse/bronze/raw_sources/engineering/source_erp/PX_CAT_G1V2.csv",
        "table": "erp_px_cat_g1v2"
    }
]
```

```python
CATALOG = "lakehouse"
SCHEMA = "bronze"


for item in INGESTION_CONFIG:

    table_name = f"{CATALOG}.{SCHEMA}.{item['table']}"

    print(f"Loading {item['source']} → {table_name}")

    df = (
        spark.read
        .format("csv")
        .option("header", True)
        .option("inferSchema", True)
        .load(item["path"])
    )

    (
        df.write
        .format("delta")
        .mode("overwrite")
        .saveAsTable(table_name)
    )

    print(f" {table_name} created")
```
